**Load the data**

```python
from csv import reader
```

```python
def load_data(dataset, num_of_rows, header=True):
    opened_file = open(dataset, encoding='utf8')
    df = list(reader(opened_file))
    df_slice = df[:num_of_rows+1]
    if header:
        return df_slice[0], df_slice[1:]
    return df_slice
```

```python
googlestore_df = load_data('googleplaystore.csv', 5000)
applestore_df = load_data('AppleStore.csv', 5000)
```

```python
def extract_data(data, index):
    column = {}
    for row in data:
        cat = row[index]
        if cat in column:
            column[cat] += 1
        else:
            column[cat] = 1   
    return column
```

```python
for row in googlestore_df_value:
    if '–' in row[0]:
        row[0] = row[0].replace('–', '-')
    elif '—' in row[0]:
        row[0] = row[0].replace('—', '-')
    elif '™' in row[0]:
        row[0] = row[0].replace('™', '')
    elif '®' in row[0]:
        row[0] = row[0].replace('®', '')
    elif '’' in row[0]:
        row[0] = row[0].replace('’', "\'")
```
