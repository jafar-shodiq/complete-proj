# Insights

```python
df1 = pd.read_csv('./CSV/canva_test_participant/participant.csv', sep=';')
df1
```

```python
df1['StatusLevelParticipant'].value_counts().sort_values().plot(kind='barh', figsize=(10, 5))
```

```python
year_dict = {}
month_dict = {}
day_dict = {}

for x in df1['LastUpdated']:
    splitted = x.split('-')
    year = splitted[0]
    month = splitted[1]
    day = splitted[2][0:2]
    
    if year not in year_dict:
        year_dict[year] = 1
    else:
        year_dict[year] += 1
        
    if month not in month_dict:
        month_dict[month] = 1
    else:
        month_dict[month] += 1
        
    if day not in day_dict:
        day_dict[day] = 1
    else:
        day_dict[day] += 1
```
