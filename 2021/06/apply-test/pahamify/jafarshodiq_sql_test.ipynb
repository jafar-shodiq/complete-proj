```python
import mysql.connector as sql
import pandas as pd
```

```python
host = 'localhost'
user = 'root'
password = ''
db_packages = 'pahamify_packages'
db_participant = 'pahamify_test_participant'
db_transaction = 'pahamify_transaction'
db_users = 'pahamify_users'
```

## 3a

```python
con = sql.connect(host=host, database=db_transaction, user=user, password=password)
cur = con.cursor()
query = 'SELECT * FROM user_status'
cur.execute(query)
table_rows = cur.fetchall()
df3a = pd.DataFrame(table_rows)
```

```python
col_names = []
for i in cur.description:
    col_names.append(i[0])

df3a.columns = col_names
df3a.set_index(col_names[0], inplace=True)
```

```python
df3a[df3a['UserPremiumStatus']==1]
```
